```python
from google.colab import drive
import pandas as pd
# Mount Google Drive
drive.mount('/content/drive')
```

```python
PATH = '/content/drive/MyDrive/GentrificAItion/Montreal_data_from_census/2011/2011 National Household Survey: Data tables/98-316-XWE2011001-401.csv'
mylist = []
for chunk in pd.read_csv(PATH, encoding='latin1', skiprows=1, chunksize=10000):
    mylist.append(chunk.loc[chunk["CMACA_Name"] == "Montréal"])

census_df = pd.concat(mylist, axis=0)
del mylist
```

```python
census_df
```

```python
census_df.to_csv("/content/drive/MyDrive/GentrificAItion/Montreal_data_from_census/2011_montreal_data.csv")
```

```python
PATH = '/content/drive/MyDrive/GentrificAItion/Montreal_data_from_census/2011/NHS_Quebec.csv'
mylist = []
for chunk in pd.read_csv(PATH, encoding='latin1', skiprows=0, chunksize=10000):
    mylist.append(chunk.loc[chunk["CMA_CA_Name"] == "Montréal"])

census_df_NHS = pd.concat(mylist, axis=0)
del mylist
census_df_NHS

census_df_NHS.to_csv("/content/drive/MyDrive/GentrificAItion/Montreal_data_from_census/2011_montreal_data_2.csv")
```
