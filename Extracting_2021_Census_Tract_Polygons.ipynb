creating geoJSON file

```python
import pandas as pd

df = pd.read_csv('/content/drive/MyDrive/Ai4Good Project/GentrificAItion/FrontEnd/2021_tract_geometry.csv')
print(df.head())
```

```python
import re


# Ensure 'geometry' column is treated as string
df['geometry'] = df['geometry'].astype(str)
print(df['geometry'])

# Function to extract coordinate pairs from POLYGON format
def extract_coordinates(polygon_str):
    coordinates = re.findall(r'[-+]?\d*\.\d+|[-+]?\d+', polygon_str)
    # Converting coordinates to floats and pairing them as (x, y)
    coordinates_pairs = [(float(coordinates[i]), float(coordinates[i+1])) for i in range(0, len(coordinates), 2)]
    # Formatting as required with square brackets
    formatted_pairs = [f"[{x}, {y}]" for x, y in coordinates_pairs]
    return formatted_pairs

# Applying the function to each row in the DataFrame
df['geometry'] = df['geometry'].apply(extract_coordinates)

# Displaying the transformed DataFrame
print(df)
```

```python
import pandas as pd
import json

def create_geojson(df):
    geojson = {
        "type": "FeatureCollection",
        "features": []
    }

    years = [2001, 2006, 2011, 2016, 2021, 2026]

    for index, row in df.iterrows():
        tract_id = row['CTNAME']
        coordinates = row['geometry']  # Ensure this is a list of lists of numerical values

        for year in years:
            feature = {
                "type": "Feature",
                "geometry": {
                    "type": "Polygon",
                    "coordinates": coordinates  # Use the coordinates directly
                },
                "properties": {
                    "tract_id": tract_id,
                    "year": year,
                    "gentrified": None,
                    "gentrifiable": None
                }
            }
            geojson["features"].append(feature)

    return geojson
```

```python
# Create the GeoJSON structure
geojson_data = create_geojson(df)

# Write the GeoJSON data to a file
with open('tracts_geojson.json', 'w') as f:
    json.dump(geojson_data, f, indent=4)
```
