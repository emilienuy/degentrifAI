```python
import geopandas as gpd
from pyproj import CRS, Transformer

# Load the shapefile
gdf = gpd.read_file('/content/lct_000a21a_e.shp')
```

```python

# Assuming the shapefile's coordinate system is not WGS84, you'll need to know its CRS
# Check the CRS of the shapefile
print(gdf.crs)

# Define the source CRS (replace with the actual CRS of your shapefile if it's not already WGS84)
source_crs = gdf.crs
```

```python

# Define the target CRS (WGS84)
target_crs = CRS("EPSG:4326")

# Initialize the transformer
transformer = Transformer.from_crs(source_crs, target_crs, always_xy=True)

```

```python
# Function to transform coordinates from source CRS to target CRS
def transform_coordinates(geometry):
    if geometry.geom_type == 'Polygon':
        coords = list(geometry.exterior.coords)
        transformed_coords = [transformer.transform(x, y) for x, y in coords]
        return transformed_coords
    elif geometry.geom_type == 'MultiPolygon':
        transformed_coords = []
        for poly in geometry:
            coords = list(poly.exterior.coords)
            transformed_coords.append([transformer.transform(x, y) for x, y in coords])
        return transformed_coords
    else:
        return None  # Handle other geometry types if needed


# Apply the transformation to all geometries
gdf['latlon_coordinates'] = gdf['geometry'].apply(transform_coordinates)
```

```python

# Save the coordinates to a CSV file or any other format
gdf[['latlon_coordinates']].to_csv('polygon_latlon_coordinates.csv', index=False)
```

```python
import re


# Ensure 'geometry' column is treated as string
df_coordinates['geometry'] = df_coordinates['geometry'].astype(str)

# Function to extract coordinate pairs from POLYGON format
def extract_coordinates(polygon_str):
    coordinates = re.findall(r'[-+]?\d*\.\d+|[-+]?\d+', polygon_str)
    # Converting coordinates to floats and pairing them as (x, y)
    coordinates_pairs = [(float(coordinates[i]), float(coordinates[i+1])) for i in range(0, len(coordinates), 2)]
    # Formatting as required with square brackets
    formatted_pairs = [f"[{x}, {y}]" for x, y in coordinates_pairs]
    return formatted_pairs

# Applying the function to each row in the DataFrame
df_coordinates['geometry'] = df_coordinates['geometry'].apply(extract_coordinates)

# Displaying the transformed DataFrame
print(df_coordinates)
```

```python

def create_geojson(df):

    geojson = {
        "type": "FeatureCollection",
        "features": []
    }

    years = [2001, 2006, 2011, 2016, 2021]

    for index, row in df.iterrows():
        tract_id = row['ctuid']
        coordinates = row['geometry']
        xycoords = []

        for item in coordinates:
          xycoords.append(item.replace('"', ''))


        for year in years:
            # Check if the year column exists in the DataFrame, else assign None
            gentrified = row.get(f'Gentrified Ding {year}', None)
            gentrifiable = row.get(f'Gentrifiable Ding {year}', None)
            gentrification_level = row.get(f'Gentrification Level Ding {year}', None)

            feature = {
                "type": "Feature",
                "geometry": {
                    "type": "Polygon",
                    "coordinates": xycoords
                },
                "properties": {
                    "tract_id": tract_id,
                    "year": year,
                    "gentrified": gentrified,
                    "gentrifiable": gentrifiable,
                    "gentrification_level": gentrification_level
                }
            }
            geojson["features"].append(feature)

    return geojson
```

```python
# Create the GeoJSON structure
geojson_data = create_geojson(df)

# Write the GeoJSON data to a file
with open('/content/drive/MyDrive/Ai4Good Project/GentrificAItion/FrontEnd/Final_data/final.geojson', 'w') as f:
    json.dump(geojson_data, f, indent=4)
```
