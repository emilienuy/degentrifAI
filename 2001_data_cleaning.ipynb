```python
import pandas as pd

df = pd.read_csv('/content/drive/MyDrive/Ai4Good Project/GentrificAItion/Montreal_data_from_census/2001_2006/2006_census_data.csv')
```

```python
# Rename the columns
df.columns = df.columns.str.replace(r'v_CA06_\d+: ', '', regex=True)

# Display the first few rows to check the new column names
print(df.head())
```

```python
# drop nulls
# Create a boolean mask where 'x' or '-' are True
mask = df.applymap(lambda x: x == 'x' or x == '-')

# Identify rows with 'x' or '-'
rows_to_drop = mask.any(axis=1)

# Drop those rows
df = df[~rows_to_drop]
```
