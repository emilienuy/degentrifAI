Calculate Gentrified Ding Measures

```python
# helper function

def weighted_quantile(values, weights, quantiles):
    """
    Compute the weighted quantiles of data.

    Parameters:
    values : array-like
        Input data.
    weights : array-like
        Weights for the data.
    quantiles : array-like
        Quantiles to compute, which should be between 0 and 1.

    Returns:
    array-like
        Weighted quantiles of the input data.
    """
    values = np.array(values)
    weights = np.array(weights)

    # Remove NaN values
    mask = ~np.isnan(values) & ~np.isnan(weights)
    values = values[mask]
    weights = weights[mask]

    # Check if there are any values left after removing NaNs
    if len(values) == 0:
        return np.nan if isinstance(quantiles, float) else [np.nan] * len(quantiles)

    # Sort values and weights
    sorter = np.argsort(values)
    values = values[sorter]
    weights = weights[sorter]

    # Compute the cumulative sum of weights
    weighted_quantiles = np.cumsum(weights) - 0.5 * weights
    weighted_quantiles /= np.sum(weights)

    # Ensure quantiles is an array
    quantiles = np.array(quantiles)

    # Interpolate to get the quantiles
    result = np.interp(quantiles, weighted_quantiles, values)

    return result

# Redefine the weighted median to use the weighted_quantile function for consistency
def weighted_median(values, weights):
    return weighted_quantile(values, weights, 0.5)
```

1 Median Home Value

```python
# Calculation: housing_price_change > housing_price_change_median

# get dwelling population for each tract
housing_price_pop_2021 = df["Total - Owner households in non-farm, non-reserve private dwellings - 25% sample data 2021"]
housing_price_pop_2016 = df["Total - Owner households in non-farm, non-reserve private dwellings - 25% sample data 2016"]
housing_price_pop_2011 = df["Number of owner households in non-farm, non-reserve private dwellings 2011"]
housing_price_pop_2006 = df["Owner-occupied private non-farm, non-reserve dwellings 2006"]
housing_price_pop_2001 = df["Owner households in non-farm, non-reserve private dwellings 2001"]

# get avg home value for each tract
housing_price_2021 = df["Average value of dwellings ($) 2021"]
housing_price_2016 = df["Average value of dwellings ($) 2016"]
housing_price_2011 = df["Average value of dwellings ($) 2011"]
housing_price_2006 = df["Average value of dwelling $ 2006"]
housing_price_2001 = df["Average value of dwelling $ 2001"]

# calculate housing_price_change
housing_price_change_2021 = (housing_price_2021 - housing_price_2016) / housing_price_2016 * 100
housing_price_change_2016 = (housing_price_2016 - housing_price_2011) / housing_price_2011 * 100
housing_price_change_2011 = (housing_price_2011 - housing_price_2006) / housing_price_2006 * 100
housing_price_change_2006 = (housing_price_2006 - housing_price_2001) / housing_price_2001 * 100


# Calculate the weighted median using the weighted_median function
housing_price_change_median_2021 = weighted_median(housing_price_change_2021, housing_price_pop_2021)
housing_price_change_median_2016 = weighted_median(housing_price_change_2016, housing_price_pop_2016)
housing_price_change_median_2011 = weighted_median(housing_price_change_2011, housing_price_pop_2011)
housing_price_change_median_2006 = weighted_median(housing_price_change_2006, housing_price_pop_2006)
```

2 Median Rent


```python
# Calculation: avg_month_rent_change > avg_month_rent_change_median

# get total renters per year
avg_rent_pop_2021 = df["Total - Tenant households in non-farm, non-reserve private dwellings - 25% sample data 2021"]
avg_rent_pop_2016 = df["Total - Tenant households in non-farm, non-reserve private dwellings - 25% sample data 2016"]
avg_rent_pop_2011 = df["Number of tenant households in non-farm, non-reserve private dwellings 2011"]
avg_rent_pop_2006 = df["Tenant-occupied private non-farm, non-reserve dwellings 2006"]
avg_rent_pop_2001 = df["Owner households in non-farm, non-reserve private dwellings 2001"]

# avg_rent per year

avg_rent_2021 = df["Median monthly shelter costs for rented dwellings ($) 2021"]
avg_rent_2016 = df["Median monthly shelter costs for rented dwellings ($) 2016"]
avg_rent_2011 = df["Median monthly shelter costs for rented dwellings ($) 2011"]
avg_rent_2006 = df["Average gross rent $ 2006"]
avg_rent_2001 = df["Average gross rent $ 2001"]

# calculate avg_rent_change
avg_rent_change_2021 = (avg_rent_2021 - avg_rent_2016) / avg_rent_2016 * 100
avg_rent_change_2016 = (avg_rent_2016 - avg_rent_2011) / avg_rent_2011 * 100
avg_rent_change_2011 = (avg_rent_2011 - avg_rent_2006) / avg_rent_2006 * 100
avg_rent_change_2006 = (avg_rent_2006 - avg_rent_2001) / avg_rent_2001 * 100

# Calculate the weighted median using the weighted_median function
avg_rent_change_median_2021 = weighted_median(avg_rent_change_2021, avg_rent_pop_2021)
avg_rent_change_median_2016 = weighted_median(avg_rent_change_2016, avg_rent_pop_2016)
avg_rent_change_median_2011 = weighted_median(avg_rent_change_2011, avg_rent_pop_2011)
avg_rent_change_median_2006 = weighted_median(avg_rent_change_2006, avg_rent_pop_2006)
```

3 Percent Higher Education Change

```python
# pct_high_educ_change > pct_high_educ_change_median

# identify column that gives population on which education is based
pct_high_educ_pop_2021 = df["Total - Highest certificate, diploma or degree for the population aged 15 years and over in private households - 25% sample data 2021"]
pct_high_educ_pop_2016 = df["Total - Highest certificate, diploma or degree for the population aged 15 years and over in private households - 25% sample data 2016"]
pct_high_educ_pop_2011 = df["Total population aged 15 years and over by highest certificate, diploma or degree 2011"]
pct_high_educ_pop_2006 = df["Total population 15 to 24 years by highest certificate, diploma or degree - 20% sample data 2006"] \
                          + df["Total population 25 to 64 years by highest certificate, diploma or degree - 20% sample data 2006"] \
                          + df["Total population 65 years and over by highest certificate, diploma or degree - 20% sample data 2006"]
pct_high_educ_pop_2001 = df['0 to 19 years 2001'] + df['20 to 29 years 2001'] \
                          + df['30 to 39 years 2001'] \
                          + df['40 to 49 years 2001'] \
                          + df['50 to 59 years 2001'] \
                          + df['60+ years 2001']

# Calculate % education per year
pct_high_educ_2021 = df["Bachelor's degree or higher 2021"] / pct_high_educ_pop_2021 * 100
pct_high_educ_2016 = df["University certificate, diploma or degree at bachelor level or above 2016"] / pct_high_educ_pop_2016 * 100
pct_high_educ_2011 = df["University certificate, diploma or degree at bachelor level or above 2011"] / pct_high_educ_pop_2011 * 100
pct_high_educ_2006 = df["University certificate or degree 2006"] / pct_high_educ_pop_2006 * 100
pct_high_educ_2001 = df["With bachelor's degree or higher 2001"] / pct_high_educ_pop_2001 * 100

# Calculate the change in percentage of high education for each year compared to its previous year
pct_high_educ_change_2021 = pct_high_educ_2021 - pct_high_educ_2016
pct_high_educ_change_2016 = pct_high_educ_2016 - pct_high_educ_2011
pct_high_educ_change_2011 = pct_high_educ_2011 - pct_high_educ_2006
pct_high_educ_change_2006 = pct_high_educ_2006 - pct_high_educ_2001

# Calculate weighted mean of changes for each year
pct_high_educ_change_median_2021 = weighted_median(pct_high_educ_change_2021, pct_high_educ_pop_2021)
pct_high_educ_change_median_2016 = weighted_median(pct_high_educ_change_2016, pct_high_educ_pop_2016)
pct_high_educ_change_median_2011 = weighted_median(pct_high_educ_change_2011, pct_high_educ_pop_2011)
pct_high_educ_change_median_2006 = weighted_median(pct_high_educ_change_2006, pct_high_educ_pop_2006)
```

Graded Gentrification Levels

```python
# Graded gentrification levels:
#  - Continued gentrification: not implemented (would require to load census data from 1996, or even from 1986)
#  - Weak gentrification: both median rent prices and home values in the bottom quartile of gentrifying CTs
#  - Intense gentrification: either median rent prices or median home values in the top quartile of gentrifying CTs
#  - Moderate gentrification: remaining CTs not categorized as weak or intense gentrification
```

```python
# "weak": housing_price_change < housing_price_change_bottomQuartile & avg_month_rent_change < avg_month_rent_change_bottomQuartile
# "intense": housing_price_change > housing_price_change_topQuartile | avg_month_rent_change > avg_month_rent_change_topQuartile
# "moderate": not weak & not intense & gentrified

# calculate quartiles for housing
housing_price_change_bottomQuartile_2021, housing_price_change_topQuartile_2021 = weighted_quantile(
    housing_price_change_2021, housing_price_pop_2021, [0.25, 0.75])
housing_price_change_topQuartile_2016, housing_price_change_bottomQuartile_2016 = weighted_quantile(
    housing_price_change_2016, housing_price_pop_2016, [0.25, 0.75])
housing_price_change_topQuartile_2011, housing_price_change_bottomQuartile_2011 = weighted_quantile(
    housing_price_change_2011, housing_price_pop_2011, [0.25, 0.75])
housing_price_change_topQuartile_2006, housing_price_change_bottomQuartile_2006 = weighted_quantile(
    housing_price_change_2006, housing_price_pop_2006, [0.25, 0.75])

# calculate quartiles for avg_month
avg_month_rent_change_bottomQuartile_2021, avg_month_rent_change_topQuartile_2021 = weighted_quantile(
        avg_rent_change_2021, avg_rent_pop_2021, [0.25, 0.75])
avg_month_rent_change_bottomQuartile_2016, avg_month_rent_change_topQuartile_2016 = weighted_quantile(
        avg_rent_change_2016, avg_rent_pop_2016, [0.25, 0.75])
avg_month_rent_change_bottomQuartile_2011, avg_month_rent_change_topQuartile_2011 = weighted_quantile(
        avg_rent_change_2011, avg_rent_pop_2011, [0.25, 0.75])
avg_month_rent_change_bottomQuartile_2006, avg_month_rent_change_topQuartile_2006 = weighted_quantile(
        avg_rent_change_2006, avg_rent_pop_2006, [0.25, 0.75])


# add weak columns
# Define weak, intense, and moderate gentrification conditions
df['Weak Gentrified Ding 2021'] = df['Gentrified Ding 2021'] \
                              & (housing_price_change_2021 < housing_price_change_bottomQuartile_2021) \
                              & (avg_rent_change_2021 < avg_month_rent_change_bottomQuartile_2021)

df['Weak Gentrified Ding 2016'] = df['Gentrified Ding 2016'] \
                              & (housing_price_change_2016 < housing_price_change_bottomQuartile_2016) \
                              & (avg_rent_change_2016 < avg_month_rent_change_bottomQuartile_2016)

df['Weak Gentrified Ding 2011'] = df['Gentrified Ding 2011'] \
                              & (housing_price_change_2011 < housing_price_change_bottomQuartile_2011) \
                              & (avg_rent_change_2011 < avg_month_rent_change_bottomQuartile_2011)

df['Weak Gentrified Ding 2006'] = df['Gentrified Ding 2006'] \
                              & (housing_price_change_2006 < housing_price_change_bottomQuartile_2006) \
                              & (avg_rent_change_2006 < avg_month_rent_change_bottomQuartile_2006)

# add high columns
df['Intense Gentrified Ding 2021'] = df['Gentrified Ding 2021'] \
                              & ((housing_price_change_2021 > housing_price_change_topQuartile_2021) \
                              | (avg_rent_change_2021 > avg_month_rent_change_topQuartile_2021))

df['Intense Gentrified Ding 2016'] = df['Gentrified Ding 2016'] \
                              & ((housing_price_change_2016 > housing_price_change_topQuartile_2016) \
                              | (avg_rent_change_2016 > avg_month_rent_change_topQuartile_2016))

df['Intense Gentrified Ding 2011'] = df['Gentrified Ding 2011'] \
                              & ((housing_price_change_2011 > housing_price_change_topQuartile_2011) \
                              | (avg_rent_change_2011 > avg_month_rent_change_topQuartile_2011))

df['Intense Gentrified Ding 2006'] = df['Gentrified Ding 2006'] \
                              & ((housing_price_change_2006 > housing_price_change_topQuartile_2006) \
                              | (avg_rent_change_2006 > avg_month_rent_change_topQuartile_2006))

# add moderate columns
df['Moderate Gentrified Ding 2021'] = df['Gentrified Ding 2021'] & ~df['Weak Gentrified Ding 2021'] & ~df['Intense Gentrified Ding 2021']
df['Moderate Gentrified Ding 2016'] = df['Gentrified Ding 2016'] & ~df['Weak Gentrified Ding 2016'] & ~df['Intense Gentrified Ding 2016']
df['Moderate Gentrified Ding 2011'] = df['Gentrified Ding 2011'] & ~df['Weak Gentrified Ding 2011'] & ~df['Intense Gentrified Ding 2011']
df['Moderate Gentrified Ding 2006'] = df['Gentrified Ding 2006'] & ~df['Weak Gentrified Ding 2006'] & ~df['Intense Gentrified Ding 2006']
```
