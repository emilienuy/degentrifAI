```python
df_2001 = pd.read_csv('/content/drive/MyDrive/GentrificAItion/Montreal_data_from_census/Standardized_Census_Data/2001_census_standardized.csv')
df_2006 = pd.read_csv('/content/drive/MyDrive/GentrificAItion/Montreal_data_from_census/Standardized_Census_Data/2006_census_standardized.csv')
df_2011 = pd.read_csv('/content/drive/MyDrive/GentrificAItion/Montreal_data_from_census/Standardized_Census_Data/2011_census_standardized.csv')
df_2016 = pd.read_csv('/content/drive/MyDrive/GentrificAItion/Montreal_data_from_census/Standardized_Census_Data/2016_census_standardized.csv')
df_2021 = pd.read_csv('/content/drive/MyDrive/GentrificAItion/Montreal_data_from_census/Standardized_Census_Data/2021_census_standardized.csv')

df_2001 = df_2001.rename(columns={'target_ctuid': 'ctuid'})
df_2006 = df_2006.rename(columns={'target_ctuid': 'ctuid'})
df_2011 = df_2011.rename(columns={'target_ctuid': 'ctuid'})
df_2016 = df_2016.rename(columns={'target_ctuid': 'ctuid'})


print(df_2021.describe())


'''
print(len(df_2006.columns))
print(len(df_2011.columns))
print(len(df_2016.columns))
print(len(df_2021.columns))
'''
```

```python
from functools import reduce

df = reduce(lambda x,y: pd.merge(x,y, on='ctuid', how='outer'), [df_2001, df_2006, df_2011, df_2016, df_2021])

print(len(df.columns))

df.describe()
```

```python
# GENTRIFIED = ((median_rent_tract < median_rent_CMA) or (median_home_value_tract < median_home_value_CMA)) and ((increase_uni_educated_residents_tract > increase_uni_educated_residents_CMA))

median_rent_col = ['Average gross rent $ 2001',
                   'Average gross rent $ 2006',
                   'Median monthly shelter costs for rented dwellings ($) 2011',
                   'Median monthly shelter costs for rented dwellings ($) 2016',
                   'Median monthly shelter costs for rented dwellings ($) 2021'
                   ]

median_home_value_col = ['Average value of dwelling $ 2001',
                         'Average value of dwelling $ 2006',
                         'Average value of dwellings ($) 2011',
                         'Average value of dwellings ($) 2016',
                         'Average value of dwellings ($) 2021'
                         ]

uni_educated_col = ["With bachelor's degree or higher 2001",
                    'University certificate or degree 2006',
                    'University certificate, diploma or degree at bachelor level or above 2011',
                    'University certificate, diploma or degree at bachelor level or above 2016',
                    "Bachelor's degree or higher 2021"
                    ]

median_rent_CMA_2021 = df.loc[df['ctuid'] == 462, "Median monthly shelter costs for rented dwellings ($) 2021"].values[0]
median_home_value_CMA_2021 = df.loc[df['ctuid'] == 462, "Average value of dwellings ($) 2021"].values[0]
uni_educated_CMA_2021 = df.loc[df['ctuid'] == 462, "Bachelor's degree or higher 2021"].values[0]

median_rent_CMA_2016 = df.loc[df['ctuid'] == 462, "Median monthly shelter costs for rented dwellings ($) 2016"].values[0]
median_home_value_CMA_2016 = df.loc[df['ctuid'] == 462, "Average value of dwellings ($) 2016"].values[0]
uni_educated_CMA_2016 = df.loc[df['ctuid'] == 462, "University certificate, diploma or degree at bachelor level or above 2016"].values[0]

median_rent_CMA_2011 = df.loc[df['ctuid'] == 462, "Median monthly shelter costs for rented dwellings ($) 2011"].values[0]
median_home_value_CMA_2011 = df.loc[df['ctuid'] == 462, "Average value of dwellings ($) 2011"].values[0]
uni_educated_CMA_2011 = df.loc[df['ctuid'] == 462, "University certificate, diploma or degree at bachelor level or above 2011"].values[0]

uni_increase_CMA_2021 = uni_educated_CMA_2021 - uni_educated_CMA_2016
uni_increase_CMA_2016 = uni_educated_CMA_2016 - uni_educated_CMA_2011

uni_educated_2021 = df["Bachelor's degree or higher 2021"]
uni_educated_2016 = df["University certificate, diploma or degree at bachelor level or above 2016"]
uni_educated_2011 = df["University certificate, diploma or degree at bachelor level or above 2011"]

uni_increase_2021 = uni_educated_2021 - uni_educated_2016
uni_increase_2016 = uni_educated_2016 - uni_educated_2011

df['Gentrified Ding 2021'] = ((df['Median monthly shelter costs for rented dwellings ($) 2021'] < median_rent_CMA_2021) \
                              | (df['Average value of dwellings ($) 2021'] < median_home_value_CMA_2021)) \
                              & (uni_increase_2021 > uni_increase_CMA_2021)

df['Gentrified Ding 2016'] = ((df['Median monthly shelter costs for rented dwellings ($) 2016'] < median_rent_CMA_2016) \
                              | (df['Average value of dwellings ($) 2016'] < median_home_value_CMA_2016)) \
                              & (uni_increase_2016 > uni_increase_CMA_2016)

df.head()
```
