```python
import pandas as pd

df = pd.read_csv('/content/drive/MyDrive/Ai4Good Colabs (Emily Hong)/Copy of 2016_montreal_data.csv')

print(df)
print(len(df))
```

```python
# Loop through questions and save the ones we want to keep

conditions = ['Population percentage change, 2011 to 2016',
              'Total private dwellings',
              'Private dwellings occupied by usual residents',
              'Population density per square kilometre',
              'Median total income in 2015 among recipients ($)',
              'Average household size', 'Non-immigrants', 'Immigrants',
              'Total visible minority population', 'Owner', 'Renter',
              'Band housing', 'Condominium', 'Not condominium',
              'Average number of rooms per dwelling', 'Suitable', 'Not suitable',
              'Only regular maintenance or minor repairs needed',
              'Major repairs needed',
              'Average monthly shelter costs for owned dwellings ($)',
              'Average value of dwellings ($)',
              '% of tenant households in subsidized housing',
              'Average monthly shelter costs for rented dwellings ($)',
              'In the labour force', 'Employed', 'Unemployed']

new_df = df.loc[df['DIM: Profile of Census Tracts (2247)'].isin(conditions)]

print(new_df)
```

```python
# drop gender columns

new_df = new_df.drop('Dim: Sex (3): Member ID: [2]: Male', axis=1)
new_df = new_df.drop('Dim: Sex (3): Member ID: [3]: Female', axis=1)
new_df = new_df.drop('Notes: Profile of Census Tracts (2247)', axis=1)
new_df = new_df.drop('Member ID: Profile of Census Tracts (2247)', axis=1)
new_df = new_df.drop('GEO_LEVEL', axis=1)
new_df = new_df.drop('GEO_CODE (POR)', axis=1)
new_df = new_df.drop('GNR', axis=1)
new_df = new_df.drop('GNR_LF', axis=1)
new_df = new_df.drop('DATA_QUALITY_FLAG', axis=1)
new_df = new_df.drop('ALT_GEO_CODE', axis=1)


print(new_df.head())
```

```python
print(len(new_df))
```

```python
# loop over rows, make a dict
# items = {'region':{}}
items = {}
for i in range(len(new_df)):
  # if i > 10:
  #   break
  # print(new_df.iloc[i]['GEO_NAME'])
  outer_key = new_df.iloc[i]['GEO_NAME']
  if outer_key in items:
    inner_key = new_df.iloc[i]['DIM: Profile of Census Tracts (2247)'] + ' 2016'
    items[outer_key][inner_key] = new_df.iloc[i]['Dim: Sex (3): Member ID: [1]: Total - Sex']
  else:
    items[outer_key] = {}
print(items)
```

```python
print(items.keys())
```

```python
# Convert the dictionary of dictionaries into a DataFrame
df = pd.DataFrame.from_dict(items, orient='index')

# Write the DataFrame to a CSV file
df.to_csv('output.csv')

print(df)
```
